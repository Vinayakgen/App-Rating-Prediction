## Data preprocessing

```python
inp1 = inp01.copy()
```

```python
inp1
```

#### 8.1 Apply log transformation (np.log1p) to Reviews and Installs.

```python
inp1['Installs']= np.log1p(inp1['Installs'])
inp1['Reviews']= np.log1p(inp1['Reviews'])
```

#### 8.2 Drop columns App, Last Updated, Current Ver, and Android Ver

```python
inp1.drop(['App','Last Updated','Current Ver','Android Ver','Type'],axis=1,inplace=True)
```

```python
inp1.head()
```

#### 8.3 Get dummy columns for Category, Genres, and Content Rating Name of dataframe should be inp2.

```python
cat_cols = ['Category', 'Genres', 'Content Rating']
inp2 = pd.get_dummies(inp1, columns=cat_cols, drop_first=True)
```

```python
inp2.head()
```

## 9. Train test split  and apply 70-30 split.

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(inp2, train_size=0.70, random_state=0)

df_train.shape,df_test.shape
```

## 10. Separate the dataframes into X_train, y_train, X_test, and y_test.

```python
y_train=df_train.Rating
X_train=df_train.drop(['Rating'],axis=1)
```

```python
X_train.shape
```

```python
y_test=df_test.Rating
X_test=df_test.drop(['Rating'],axis=1)
```

```python
X_test.shape
```

## 11. Model building

```python
from sklearn.linear_model import LinearRegression
reg =LinearRegression()
reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
y_train_pred= reg.predict(X_train)
r2_score(y_train, y_train_pred)
```

## 12. predictions on test set and report R2.

```python
y_test_pred = reg.predict(X_test)
```

```python
dfReg = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})
```

```python
dfReg
```

```python
y_test_pred= reg.predict(X_test)
r2_score(y_test, y_test_pred)
```

```python
from sklearn.metrics import mean_squared_error as ms
print('Root_Mean_Squared_Error(RMSE)=',np.sqrt(ms(y_test,y_test_pred)))
```
